```python
import numpy as np
import pandas as pd
```

```python
#definir la funcion
def CrearDataSet(Num=1):
    
    Output = []
    
    for i in range(Num):
        
        # Crear un rango de fechas semanal (De lunes a lunes)
        rng = pd.date_range(start='1/1/2016', end='12/31/2020', freq='W-MON') #frecuencia semanal
        
        # Crear valores aleatorios
        data = np.random.randint(low=25,high=1000,size=len(rng)) #hasta el numero total(len) de fechas semanales que tenemos (las cuenta)
        
        # Estatus posibles
        status = [1,2,3]
        
        # Lista de estatus aleatorios
        random_status = [status[np.random.randint(low=0,high=len(status))] for i in range(len(rng))] #rango entre 0 y el contado de estados, para el rango,(cantidad de fechas optenidas)
     
        # Locales posibles
        locales = ['Libertador','El Hatillo','El hatillo','Chacao','Baruta','Sucre'] #nombramos los locales
       
        # Crear una lista aleatoria de estatuses
        random_locales = [locales[np.random.randint(low=0,high=len(status))] for i in range(len(rng))] #hacer lo mismo que con los estados
        
        Output.extend(zip(random_locales, random_status, data, rng)) #(zip) itera sobre cada una de las listas desde el indice 0 de cada una para formar la lista
        
    return Output
```

```python
dataset = CrearDataSet(4) #llamamos a la funcion y le asignamos 4 

df = pd.DataFrame(data=dataset, 
                  columns=['Local','Estatus_local','Cantidad_Clientes','Fecha_Status'])
df
```

##  <font color=#003d5c> _np.random.choice()_ </font>
### Esta función extrae una muestra aleatoria a partir de un array unidimensional. 

```python
filas = np.random.choice(df.index, 10, replace=False) 
#np.random.choice para seleccionar 10 índices aleatorios del DataFrame
#replace=False: evita que se seleccionen índices repetidos
filas
```

```python
df.loc[filas] #selecciona las filas que hemos cogido y crea el dataframe
```
